It is used to find the local or global optima (maxima/minima) of a function.

y = f(x)

- Linear Regression
- Logistic Regression
- (Deep) Neural Networks
- Back-propagation algorithms

A function can have local maxima or minima as well as global maxima or minima.

## Average Rate of Change iver an interval

The average rate of y = f(x) over an interval [x1, x2] is given by:

$\frac{\Delta y}{\Delta x} = \frac{f(x_2) - f(x_1)}{x_2 - x_1}$

$x2 = x1 + h$, where $h = \Delta x$ and $h \neq 0$

Therefore,

$\frac{\Delta y}{\Delta x} = \frac{f(x_1 + h) - f(x_1)}{h}$
